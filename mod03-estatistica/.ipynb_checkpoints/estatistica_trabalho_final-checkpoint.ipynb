## 5.1 Avaliação inicial


O primeiro passo para termos uma avaliação eficiente de nossa base de dados é entender quais são as informações presentes na mesma. Para isso, normalmente fazemos uso de dois métodos que fazem parte do pandas: dataframe.head(), que será utilizado para vermos uma parte dos dados e dataframe.info(), que irá nos dar informações sobre as variáveis presentes.

```python
df_houses.head(20)
```

```python
df_houses.info()
```

Como podemos ver, temos um total de 13 variáveis (que são as colunas de nossa base de dados), sendo 9 variáveis inteiras e 4 variáveis 'objeto' (neste caso, strings), e todas com valores preenchidos. Por motivos de organização, o ideal é criar duas listas para separar estas variáveis. Isso vai ser importante para etapas futuras de nossa análise.

Antes disso, porém, é importante normalizar os nomes das colunas, de modo a facilitar o acesso às mesmas.

```python
df_houses.rename(columns = {"parking spaces": "parkingSpaces", 
                            "hoa (R$)": "hoa", 
                            "rent amount (R$)": "rentAmount", 
                            "property tax (R$)": "propertyTax", 
                            "fire insurance (R$)": "fireInsurance", 
                            "total (R$)": "total"},
                 inplace=True)
df_houses.head()
```

Agora, com os nomes em uma forma mais "amigável", faremos a separação das variáveis.

```python
colunas_numericas = [x for x in df_houses if isinstance(df_houses[x][0], np.int64)]
colunas_categoricas = [x for x in df_houses if x not in colunas_numericas]

print("Variáveis numéricas: ", colunas_numericas)
print("Variáveis categóricas: ", colunas_categoricas)
```

Olhando o head da nossa base, podemos entender o porquê das variáveis objeto: Elas são referentes à classificações (e.g.: variável "furniture", que classifica os registros entre casas decoradas e não-decoradas) ou nomes (e.g.: variável city"). 
Entretanto, há uma coluna ("floor") que parece indicar em que andar encontra-se a casa (consideremos aqui que a "casa" pode tratar-se de um apartamento). Logo, seria de se esperar que fosse uma coluna com valores do tipo inteiro. Porém, a mesma é classificada como "object". Vamos entender o motivo:

```python
df_houses.floor.unique()
```

O comando utilizado na célula acima exibe todos os valores presentes na coluna "floor", de forma única. Com isso, vemos que há um valor representado por '-', um traço. Por conta desse valor, a coluna foi interpretada como "object". A solução para esse problema seria apenas remover as entradas com esse valor. Entretanto, é importante avaliar o impacto de tomar essa atitude. Para isso, vamos avaliar quantas vezes esse valor aparece em nossa base:

```python
list(df_houses.floor).count('-')
```

Sabendo que temos um total de 10692 entradas em nossa base (da nossa chamada ao método info()), as entradas com o valor '-' representam cerca de 20% do total. Logo, seria um erro remover essas entradas.
Baseado em nossa interpretações acerca da base, inferimos que o valor '-' represente, de fato, casas e não apartamentos. Logo, o símbolo seria a representação do "Térreo". De forma a otimizar a organização de nossa base, iremos substituir o valor '-' por '0' e, em seguida, converter os valores em inteiros.

```python
df_houses.floor = [int(x) if x != '-' else 0 for x in df_houses.floor]
```

```python
df_houses.floor.unique()
```

Agora que organizamos nossa coluna "floor", seria interessante repetir esse raciocínio para as outras colunas com valores categóricos. Afinal, precisamos garantir que estas não possuam valores que possam ser problemáticos à nossa análise. 
Porém, precisamos primeiro remover a coluna "floor" da lista de colunas categóricas e colocá-la na lista de colunas numéricas. Para isso, vamos fazer da forma mais simples: repetir o mesmo processo.

```python
colunas_numericas = [x for x in df_houses if isinstance(df_houses[x][0], np.int64)]
colunas_categoricas = [x for x in df_houses if x not in colunas_numericas]

print("Variáveis numéricas: ", colunas_numericas)
print("Variáveis categóricas: ", colunas_categoricas)
```

Com nossas listas atualizadas, vamos continuar nossa verificação.

```python
for column in df_houses:
  if column in colunas_categoricas:
    print("{}: {}".format(column, df_houses[column].unique()))
```

Pelos valores exibidos, podemos ficar tranquilos quanto às nossas variáveis categóricas.

```python
df = pd.DataFrame({"columns": [column for column in df_houses 
                               if column in colunas_numericas], 
                   "Null values": [df_houses[column].isnull().values.any() 
                   for column in df_houses if column in colunas_numericas]})
df
```

## 5.2 Extraindo informações da base

Para auxiliar em alguns cálculos estatísticos, vamos criar duas funções.

A função **cv** realiza o cálculo do coeficiente de variação e a função **statistics** retorna o cálculo da média, mediana, moda, desvio-padrão, coeficiente de variação e simetria do conjunto de dados fornecido - no nosso caso, de uma variável quantitativa.

```python
def cv(n):
  num = 100*st.stdev(n)/st.mean(n)
  return round(num,2)

def statistics(df, field):
  return {
      'média': st.mean(df[field]),
      'mediana': st.median(df[field]),
      #'moda': st.multimode(df[field]),
      'coef. de variação': cv(df[field]),
      'simetria': stats.skew(df[field])
  }
```

Vamos utilizar estas funções para criar um dataframe com estas informações para todas as variáveis quatitativas do nosso dataset.

```python
stat_dict = {
    column: statistics(df_houses, column) for column in colunas_numericas
}
df_statistics = pd.DataFrame(stat_dict)
df_statistics
```

```python
fig, ax = plt.subplots()
stats.probplot(df_houses['propertyTax'], fit=True, plot=ax)
plt.show()
```

Como é possível ver, as variáveis disponíveis nesta base de dados são, em sua maioria, numéricas. Por esse motivo, temos informações bastante dispersas para a apresentação. Portanto, iremos simplificar essas informações em um dataframe mais simples.

```python
df_houses.rename(columns = {"parking spaces": "parkingSpaces", 
                            "hoa (R$)": "hoa", 
                            "rent amount (R$)": "rentAmount", 
                            "property tax (R$)": "propertyTax", 
                            "fire insurance (R$)": "fireInsurance", 
                            "total (R$)": "total"},
                 inplace=True)
df_houses.head()
```

```python
cidades = df_houses.city.unique()
data_sp = {'houses': 0,
        'area': 0,
        'rooms': 0,
        'bathrooms': 0,
        'parkingSpaces': 0,
        'hoa': 0,
        'rentAmount': 0,
        'propertyTax': 0,
        'fireInsurance': 0,
        'total': 0}
data_poa = {'houses': 0,
        'area': 0,
        'rooms': 0,
        'bathrooms': 0,
        'parkingSpaces': 0,
        'hoa': 0,
        'rentAmount': 0,
        'propertyTax': 0,
        'fireInsurance': 0,
        'total': 0}
data_rj = {'houses': 0,
        'area': 0,
        'rooms': 0,
        'bathrooms': 0,
        'parkingSpaces': 0,
        'hoa': 0,
        'rentAmount': 0,
        'propertyTax': 0,
        'fireInsurance': 0,
        'total': 0}
data_cp = {'houses': 0,
        'area': 0,
        'rooms': 0,
        'bathrooms': 0,
        'parkingSpaces': 0,
        'hoa': 0,
        'rentAmount': 0,
        'propertyTax': 0,
        'fireInsurance': 0,
        'total': 0}
data_bh = {'houses': 0,
        'area': 0,
        'rooms': 0,
        'bathrooms': 0,
        'parkingSpaces': 0,
        'hoa': 0,
        'rentAmount': 0,
        'propertyTax': 0,
        'fireInsurance': 0,
        'total': 0}

data = {'São Paulo': data_sp, 
        'Porto Alegre': data_poa, 
        'Rio de Janeiro': data_rj, 
        'Campinas': data_cp, 
        'Belo Horizonte': data_bh}
for cidade in data:
  data[cidade]['houses'] = int(df_houses.city.count())
  data[cidade]['area'] = "{:.2f}".format(round(df_houses.loc[df_houses["city"] == cidade].area.mean(),2))
  data[cidade]['rooms'] = round(df_houses.loc[df_houses["city"] == cidade].rooms.mean())
  data[cidade]['bathrooms'] = round(df_houses.loc[df_houses["city"] == cidade].bathroom.mean())
  data[cidade]['parkingSpaces'] = round(df_houses.loc[df_houses["city"] == cidade].parkingSpaces.mean())
  # data[cidade]['floor'] = sum(df_houses.loc[df_houses["city"] == cidade].floor)
  data[cidade]['hoa'] = "{:.2f}".format(round(df_houses.loc[df_houses["city"] == cidade].hoa.mean(),2))
  data[cidade]['rentAmount'] = "{:.2f}".format(round(df_houses.loc[df_houses["city"] == cidade].rentAmount.mean(),2))
  data[cidade]['propertyTax'] = "{:.2f}".format(round(df_houses.loc[df_houses["city"] == cidade].propertyTax.mean(),2))
  data[cidade]['fireInsurance'] = "{:.2f}".format(round(df_houses.loc[df_houses["city"] == cidade].fireInsurance.mean(),2))
  data[cidade]['total'] = "{:.2f}".format(round(df_houses.loc[df_houses["city"] == cidade].total.mean(),2))

df = pd.DataFrame(data).T #, index = data.keys())
df 
```
